`KNNImputer`

Impute k-nearest neigbor using a weighted or unweighted mean.

```python
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

df = pd.read_csv("../data/penguins.csv")

# mean values
before_imputation = [
    df["bill_length_mm"].mean(),
    df["bill_depth_mm"].mean(),
    df["flipper_length_mm"].mean(),
    df["body_mass_g"].mean(),
    df["delta_15_n"].mean(),
    df["delta_13_c"].mean(),
]

imputer = KNNImputer(n_neighbors=3)
df[
    [
        "bill_length_mm",
        "bill_depth_mm",
        "flipper_length_mm",
        "body_mass_g",
        "delta_15_n",
        "delta_13_c",
    ]
] = imputer.fit_transform(
    df[
        [
            "bill_length_mm",
            "bill_depth_mm",
            "flipper_length_mm",
            "body_mass_g",
            "delta_15_n",
            "delta_13_c",
        ]
    ]
)

# mean values
after_imputation = [
    df["bill_length_mm"].mean(),
    df["bill_depth_mm"].mean(),
    df["flipper_length_mm"].mean(),
    df["body_mass_g"].mean(),
    df["delta_15_n"].mean(),
    df["delta_13_c"].mean(),
]

print("Before -> After Imputation")
print("=" * 30)
[
    f"{before_imputation[i]} -> {after_imputation[i]}"
    for i in range(len(before_imputation))
]

imputer = KNNImputer(n_neighbors=2)
dataframe = pd.DataFrame(
    {
        "A": [1.0, 2.0, np.nan, 1.0],
        "B": [np.nan, 2.0, 1.0, 2.0],
        "C": [1.0, np.nan, np.nan, 2.0],
    }
)
dataframe[["A", "B", "C"]] = imputer.fit_transform(dataframe[["A", "B", "C"]])
dataframe.head()
```
